# Analysis of Data for N-Prime Numbers

```python
# Saving the data to a CSV file (Optional)
import pandas as pd
import matplotlib.pyplot as plt
```

## Table Representation
The table below shows the data for different cluster sizes, along with their corresponding 'Total' and 'Avg' values.


```python
# Load the data from the CSV file
df = pd.read_csv('results.csv')

# Display the DataFrame
df
```

## Visualization

### 1. Bar Chart: Total Time vs Clusters Size
The bar chart below visualizes the 'Total' value for each cluster size.


```python
plt.figure(figsize=(10, 6))
plt.plot(df['Clusters'], df['Total'], color='blue', marker='o', markerfacecolor='black' )
plt.xlabel('Clusters Size')
plt.ylabel('Total Time')
plt.title('Total Time vs Clusters Size')
plt.xticks(df['Clusters'])
plt.grid(True)
plt.show()
```

### 2. Line Chart: Avg vs Clusters
The line chart below shows the 'Avg' value corresponding to each cluster size.

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Clusters'], df['Avg'], marker='o', color='orange')
plt.xlabel('Clusters Size')
plt.ylabel('Average Time')
plt.title('Average Time vs Clusters Size')
plt.xticks(df['Clusters'])
plt.grid(True)
plt.show()
```
